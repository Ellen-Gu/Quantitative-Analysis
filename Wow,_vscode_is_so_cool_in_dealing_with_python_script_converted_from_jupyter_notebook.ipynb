I often think to convert my jupyter notebook files to python scripts. This can be done by expoting the notebook to exectue script in jupyter lab/notebook/hub. It can also be done by nbconvert command. The covertion is generally not an issue.

However, after convertion there will be some small modifications from time to time. Then I want the convenience which Jupter inteface provides for debugging and partially run, but at the same time I do not want to convert the whole script back into notebook.

Vscode is so nicely considering this. It add the jupyter cell alike comments in the converted python script, and provide the corresponding IDE functions to let user run a cell just like running in jupyter notebook. I take the py torch's quickstart_tutorial.ipynb and converted it to python code in jupyterhub. The following graph shows that the converted python script is not only running smoothly in terminal, but also can be run in interactive windows for testing and debugging. User only need to click the "Run Cell" button which provided by Vscode to pop up the Interactive window and run cells there, and dodebugging code at the bottom of the interactive window. This is cool. For pure python code add the # In[ ]: also do the trick, makes a pure python script run in a flavor like jupyter notebook. With the IntelliCode, python combo, Remote combo, jupter combo and this cell IDE functions, etc, Vscode does provide a very comfort code developing environment.
